```python
import numpy as np

W, H = 12, 8
img = np.full((H, W, 3), 255, dtype=np.uint8)  # RGB, all white

def set_black(r, c):
    img[r, c] = (0, 0, 0)

# ----- Draw L in left 6x8 block (cols 0..5) -----
# Vertical stroke at col 0 (rows 0..7)
for r in range(H):
    set_black(r, 0)
# Bottom stroke at row 7, cols 0..4 (leave col 5 white for spacing)
for c in range(0, 5):
    set_black(H - 1, c)

# ----- Draw P in right 6x8 block (cols 6..11) with margins -----
off = 6
stem_c = off + 1        # col 7 (left margin at col 6)
right_c = off + 4       # col 10 (right margin at col 11)

# Stem
for r in range(H):
    set_black(r, stem_c)

# Top bar (row 0), cols 7..10
for c in range(stem_c, right_c + 1):
    set_black(0, c)

# Middle bar (row 3), cols 7..10
for c in range(stem_c, right_c + 1):
    set_black(3, c)

# Right vertical for upper bowl (rows 1..2) at col 10
for r in range(1, 3):
    set_black(r, right_c)

# ----- Write P3 ASCII PPM -----
out_file = "nameinitial-LP.ppm"
with open(out_file, "w") as f:
    f.write("P3\n")
    f.write(f"{W} {H}\n")
    f.write("255\n")
    for r in range(H):
        for c in range(W):
            R, G, B = img[r, c]
            f.write(f"{R} {G} {B}  ")
        f.write("\n")

print(f"Wrote {out_file}")
```
